```python
import random
from random import choice
from collections import Counter
import json


class Menace:

    def __init__(self):
        self.board = [" "]*9
        self.beads = [10]*9
        # self.matchboxes = {}
        self.movesplayed = []
        try:
            a_file = open("data.json", "r")
            # output = a_file.read()
            self.matchboxes = json.loads(a_file.read())
        except:
            self.matchboxes = {}
            # print("No Pre Game exist")

    def PRINT_THE_BOARD(self):
        print("\n CURRENT Positions:")
        print(" 0 | 1 | 2   ", self.board[0],
              "|", self.board[1], "|", self.board[2])
        print("___+___+___    ___+___+___ ")
        print(" 3 | 4 | 5   ", self.board[3],
              "|", self.board[4], "|", self.board[5])
        print("___+___+___     ___+___+___ ")
        print(" 6 | 7 | 8   ", self.board[6],
              "|", self.board[7], "|", self.board[8])
        print("\n")

    def USER_CHANCE(self):
        pos = int(input("Enter position: "))
        if self.board[pos] != " ":
            print("Wrong input enter again")
            self.USER_CHANCE()
        else:
            self.board[pos] = "X"

    def COMPUTER_CHANCE(self):
        current_board = self.board_string()
        # print("board", movesplayed)
        if current_board not in self.matchboxes:
            new_beads = [pos for pos, mark in enumerate(
                current_board) if mark == ' ']
            # Early boards start with more beads
            self.matchboxes[current_board] = new_beads * \
                ((len(new_beads) + 2) // 2)

        current_beads = self.matchboxes[current_board]
        if len(current_beads):
            bead = random.choice(current_beads)
            self.movesplayed.append((current_board, bead))
        else:
            bead = -1
        return bead
      

    def winning(self):
        if (self.board[0] != ' ' and
            ((self.board[0] == self.board[1] == self.board[2]) or
            (self.board[0] == self.board[3] == self.board[6]) or
            (self.board[0] == self.board[4] == self.board[8]))) or (self.board[4] != ' ' and
            ((self.board[1] == self.board[4] == self.board[7]) or
            (self.board[3] == self.board[4] == self.board[5]) or
            (self.board[2] == self.board[4] == self.board[6]))) or (self.board[8] != ' ' and
            ((self.board[2] == self.board[5] == self.board[8]) or
            (self.board[6] == self.board[7] == self.board[8]))):
            return True
        else:
            return False



    def MATCH_TIED(self):
        c = Counter(self.board)
        for i in c:
            if (i == " "):
                return False
        return True



    def change_of_bead(self, n):
        if n == 3:
            for (board, bead) in self.movesplayed:
                self.matchboxes[board].extend([bead, bead, bead])
            # self.num_win += 1
        if n == 2:
            for (board, bead) in self.movesplayed:
                self.matchboxes[board].append(bead)
        if n == 1:
            # Lose, remove a bead
            for (board, bead) in self.movesplayed:
                matchbox = self.matchboxes[board]
                del matchbox[matchbox.index(bead)]



    def GAME_RESET(self):
        self.board = [" "]*9
        self.movesplayed = []



    def board_string(self):
        return ''.join(self.board)



    def START_THE_GAME(self):
        chance = 0
        while (True):
            chance += 1
            move = self.COMPUTER_CHANCE()
            self.board[move] = "O"
            # if chance>=2:
            if self.winning():
                self.PRINT_THE_BOARD()
                self.change_of_bead(3)
                print("Computer Won")
                break
            if self.MATCH_TIED():
                self.PRINT_THE_BOARD()
                self.change_of_bead(2)
                print("Game MATCH_TIED")
                break
            self.PRINT_THE_BOARD()
            self.USER_CHANCE()
            # if chance>=2:
            if self.winning():
                self.PRINT_THE_BOARD()
                self.change_of_bead(1)
                print("You Won")
                break

    def RULES_OF_GAME(self):
        print("\n")
        print("LET US PLAY MENACE TICK TACK TOE GAME ") 
        print(" USER's symbol is  X and that of Computer is O ")
        print("Start Playing!")
        print("\n")

    def EXIT_THE_GAME(self):
        ans = input("\nWanna Go ahead with the game? Yes or No \n")
        if ans.lower() == "no":
            print("Thank you for playing!! \n")
            a_file = open("data.json", "w")
            json.dump(self.matchboxes, a_file)
            a_file.close()
            return True
        else:
            a_file = open("data.json", "w")
            json.dump(self.matchboxes, a_file)
            a_file.close()
            return False

if __name__ == '__main__':
    stop = False
    MENACE_GAME = Menace()
    while(not stop):
        MENACE_GAME.RULES_OF_GAME()
        MENACE_GAME.START_THE_GAME()
        stop = MENACE_GAME.EXIT_THE_GAME()
        MENACE_GAME.GAME_RESET()
```
